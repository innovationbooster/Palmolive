# Brazil

```python
reviews_amazon_brazil.head(2)
```

```python
from googletrans import Translator, constants
translator = Translator()
```

```python
reviews_mercado_brazil_list = list(reviews_mercado_brazil['Review'])
```

```python
reviews_amazon_brazil_list = []
for i in range(len(reviews_amazon_brazil)):
    content = translator.translate(reviews_amazon_brazil['content'][i]).text
    if not isinstance(reviews_amazon_brazil['title'][i], float):
        title = translator.translate(reviews_amazon_brazil['title'][i]).text
        reviews_amazon_brazil_list.append(title + " " + content)
    else:
        reviews_amazon_brazil_list.append(content)
```

```python
reviews_mercado_brazil_list
```

```python
reviews_amazon_brazil_list
```

```python
brazil_review = reviews_mercado_brazil_list + reviews_amazon_brazil_list
```

<li> Noun Chunk Analysis </li>

```python
delete_words = ['soap', 'grams', 'units', 'gramas', 'barra', 'male', 'mercado', 'pm', 'skin', 'full', 'w3', 'product', 'zip', 'loccitane', 'shopping', 'service', 'use', 'purchase', 'shipping','great']
brazil_review_noun_chunk = nlpshortcut.chunk_generator(brazil_review, delete_words, bubbleplot = True)
```

```python
brazil_review_noun_chunk
```

<li> Pos Tag Analysis </li>

```python
adj_delete = ['free','daily','best','always','also','new','high','good','excellent','wonderful','well','really','much','long','pleasant','nice','super','already','even','first','perfect']
verb_delete = ['helps','made','contains','leaving','used','apply','indicated','g','provides','keeps','leaves','rinse','using','leave','provide','developd','make','help','contain','take','want','skin','recommended','give','achieve','gives','feeling','thank','loved','bought','recommend','liked','came']
noun_delete = ['g','products','bar','information','skin','body','foams','soap','formula','ingredients','day','water','action','rinse','bath','addition','properties','invoice','product','hair','time','purchase','hands','soaps','recommend','love']
brazil_review_pos = nlpshortcut.pos_viz(brazil_review,delete_words, adj_delete, verb_delete, noun_delete, printcount = True, topcount = 10, noun_size = 0.6, verb_size = 0.9, adj_size = 0.2)
```

<li> Lda Analysis </li>

```python
stopwords = list(set(delete_words + adj_delete + verb_delete + noun_delete)) + ['produto', 'muito', 'excelente', 'ótimo', 'will', 'pele', 'para']
LDA_analysis = nlpshortcut.ldacreator(' '.join(brazil_review),topics = 5,stopwords = stopwords, passes = 50)
```

```python
LDA_analysis
```
