```python
def get_info(prod_url):
    prod_dict = dict()
    driver.get(prod_url)
    time.sleep(3)
    try:
        name = driver.find_elements_by_xpath("//h1[@class='productName_title']")[0].text
        prod_dict['name'] = name
    except:
        pass
    try:
        price = float(driver.find_elements_by_xpath("//p[@class='productPrice_price']")[0].text.strip().split('$')[1])
        prod_dict['price'] = price
    except:
        pass
    try:
        description = driver.find_elements_by_xpath("//div[@class='productDescription_contentPropertyListItem']/div")[0].text
        prod_dict['description'] = description
    except:
        pass
    try:
        dict_ = dict()
        for item in driver.find_elements_by_xpath("//div[@class='productDescription_contentWrapper']"):
            if item.find_elements_by_xpath(".//div[@class='productDescription_contentPropertyName']")[0].text:
                dict_[item.find_elements_by_xpath(".//div[@class='productDescription_contentPropertyName']")[0].text.strip().replace(':','')] = item.find_elements_by_xpath(".//div[@class='productDescription_contentPropertyValue']")[0].text.strip()
        prod_dict.update(dict_)
    except:
        pass
    try:
        reviews = []
        for review in driver.find_elements_by_xpath("//div[@class='productReviews_topReviewSingle']"):
            review_content = review.find_elements_by_xpath(".//p")[0].text
            try:
                review_title = review.find_elements_by_xpath(".//div/h3")[0].text
                reviews.append(review_title + ' ' + review_content)
            except:
                reviews.append(review_content)
        prod_dict['reviews'] = reviews
    except:
        pass
    return prod_dict
```

```python
def get_info(prod_url):
    prod_dict = dict()
    driver.get(prod_url)
    time.sleep(3)
    try:
        name = driver.find_elements_by_xpath("//h1")[0].text.strip()
        prod_dict['name'] = name
    except:
        pass
    try:
        price = float(driver.find_elements_by_xpath("//span[@id='current-price-string']")[0].text)
        prod_dict['price'] = price
    except:
        pass
    try:
        small_desp = driver.find_elements_by_xpath("//div[@id='product-page-selling-statement']")[0].text.strip()
        prod_dict['small_desp'] = small_desp
    except:
        pass
    try:
        desp = driver.find_elements_by_xpath("//div[@class='_3LvFj']/div")[1].text
        prod_dict['desp'] = desp
    except:
        pass
    try:
        ingredients = driver.find_elements_by_xpath("//div[@class='_3LvFj']/div[@class='_2Tumz']/p")[0].text
        prod_dict['ingredients'] = ingredients
    except:
        pass
    return prod_dict   
```

```python
def get_review(prod_url):
    driver.get(prod_url)
    time.sleep(3)
    try:
        name = driver.find_elements_by_xpath("//h1")[0].text.strip()
        time.sleep(5)
        driver.find_elements_by_xpath("//div[@aria-label='dropdown']")[-1].click()
        driver.find_elements_by_xpath("//div[@class='lditi']/div/div/div/div/ul/li")[1].click()
        review_title = []
        review_content = []
        while len(review_content) <= 100:
            for title in driver.find_elements_by_xpath("//strong[@class='_3P2YP']"):
                try:
                    review_title.append(title.text.strip())
                except:
                    review_title.append('')
            for content in driver.find_elements_by_xpath("//div[@class='_2wSBV']"):
                try:
                    review_content.append(content.text.strip())
                except:
                    review_content.append('')
            try:
                if driver.find_elements_by_xpath("//a[@class='_2a4bF']")[-1].text == 'Next':
                    driver.find_elements_by_xpath("//a[@class='_2a4bF']")[-1].click()
                    time.sleep(3)
                else:
                    break
            except:
                break
        review_name = [name] * len(review_content)
        return review_name, review_title, review_content
    except:
        pass
```
